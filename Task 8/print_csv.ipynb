```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  


data = pd.read_csv('data_task8.csv')
epoch = data["epoch"]


accuracy = data["accuracy"]
val_accuracy = data["val_accuracy"]


fig = plt.figure()  
#plt.plot(epoch, val_dice_coef, label="val_dice_coef")  
plt.plot(epoch, accuracy, label="accuracy")  
plt.plot(epoch, val_accuracy, label="val_accuracy")  
plt.grid()  
plt.xlabel("epoch")  
plt.legend(loc="best")  
plt.suptitle("Do thi accuracy")  
plt.show()
```

```python
loss = data["loss"]
val_loss = data["val_loss"]


fig = plt.figure()  
#plt.plot(epoch, val_dice_coef, label="val_dice_coef")  
plt.plot(epoch, loss, label="loss")  
plt.plot(epoch, val_loss, label="val_loss")  
plt.grid()  
plt.xlabel("epoch")  
plt.legend(loc="best")  
plt.suptitle("Do thi loss")  
plt.show()
```
